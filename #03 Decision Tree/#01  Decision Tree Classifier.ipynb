### Post Preprunning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
dataset.keys()
```

```python
df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['target'] = dataset.target
```

```python
df.head()
```

```python
dataset.target
```

```python
## Independent and Dependent features
X = df.iloc[:,:-1]
y = dataset.target
```

```python
X
```

```python
y
```

```python
X,y
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

#### Post Purning

```python
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train, y_train)
```

```python
X_train.head()
```

```python
from sklearn import tree 
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier, filled = True)
```

```python
#### Post Pruning with max_dept = 2
treeclassifier = DecisionTreeClassifier(criterion='entropy', max_depth=2)
treeclassifier.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier, filled = True)
```

#### Prediction

```python
y_pred = treeclassifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_pred, y_test)
print(score)
print(classification_report(y_pred,y_test))
```

### Pre Pruning And Hypermarameter Tuning For Hug Data

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
parameter={
  'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treeclassifier = DecisionTreeClassifier()
clf = GridSearchCV(treeclassifier,param_grid=parameter,cv=5,scoring="accuracy")
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
y_pred = clf.predict(X_test)
```

```python
score = accuracy_score(y_pred,y_test)
score
```

```python
print(classification_report(y_pred,y_test))
```
